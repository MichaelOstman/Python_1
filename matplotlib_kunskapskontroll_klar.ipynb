### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
Either using object-oriented style or pyplot. The object oriented style is recommended.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure: Think of it as the entire visual space, your canvas where the plot is displayed.
Axes: This is the specific area of the canvas where your plot is. The plot can be customized by adjusting the axes.
Axis: Responsible for defining the scale, limits, and generating ticks as well as tick labels.
Artist: Any visible element within the figure, including the figure itself, axes, and their components, is considered an artist.  
```

### When plotting in Matplotlib, what is the expected input data type?

```python
Plotting functions expect numpy.array or objects that can be passed to numpy.asarray.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
#Object-oriented approach

x = np.linspace(-4, 4, 100)
y = x**2 

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('y = x^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
#pyplot approach

plt.plot(x, y)
plt.title("y = x^2")
plt.xlabel('x')
plt.ylabel('y')
```

```python
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, axs = plt.subplots(1, 2,layout='constrained')
fig.suptitle('Different Plots', size = 50)

axs[1].bar(names, values)
axs[1].set_title('Bar Plot')

axs[0].scatter(random_data_x, random_data_y)
axs[0].set_title('Scatter Plot')
```
